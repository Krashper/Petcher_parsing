```python
import pandas as pd
import re
```

```python
dataset = pd.read_csv("..\dags\data\Airflow_dataset.csv")
```

```python
dataset
```

```python
dataset = dataset.fillna("[]")

def convert_string_to_list(string):
    # Удаляем квадратные скобки
    string = string.strip("[]")
    # Разделяем строку по шаблону, который учитывает пробелы и переводы строки
    tags_list = re.split(r"\s+'\s*", string)
    # Удаляем пустые строки и очищаем от лишних символов
    cleared_tags_list = [tag.strip().replace("'", "") for tag in tags_list if tag.strip()]
    return cleared_tags_list
```

```python
dataset['Tags'] = dataset['Tags'].apply(convert_string_to_list)

dataset["Tags"] = dataset["Tags"].map(lambda x: x if isinstance(x, list) else [])
```

```python
top_tags = pd.read_csv("..\dags\data\Top_tags.csv")["Skill"].to_list()

top_tags
```

```python
dataset["Tags"] = dataset["Tags"].apply(lambda x: [element for element in x if element in top_tags])
```

```python
dataset
```

```python
def has_matching_tag(tags, top_tags):
    return any(tag in top_tags for tag in tags)

# Применяем фильтрацию
dataset = dataset[dataset['Tags'].apply(lambda x: has_matching_tag(x, top_tags))]
dataset.reset_index(drop=True, inplace=True)
```

```python
user_skill_matrix = pd.DataFrame(columns=top_tags)

for index, row in dataset.iterrows():
    tags = row["Tags"]
    for tag in tags:
        if tag in top_tags:
            user_skill_matrix.loc[index, tag] = 1
    print(f"{index}/{len(dataset)}")

user_skill_matrix
```
